```python
import pandas as pd
import numpy as np
import requests
from lxml import html # pip install lxml
```

```python
headers = {
    'USER_AGENT': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Ubuntu Chromium/71.0.3578.80 Chrome/71.0.3578.80 Safari/537.36',
    'FEED_EXPORT_ENCODING': 'utf-8' # ocultando la dirección, que es una programa el que esta visitando las URL    
  }

lugar = []
destino = []
detalles = []

try:
    for i in range(0,13):
        for j in range(15,106,15):
            url = "https://www.guatevalley.com/deportes_extremos_aventura.php?lat=14.6426619&longi=-90.513341&type=10{i}&rubro=1&name=h&num={j}".format(i=i, j=j)
            response = requests.get(url, headers = headers) # ibtiene todas las url de la lista cargada
            parser = html.fromstring(response.content)
            nombre = parser.xpath("//div//div//h4//strong/text()")
            ubicacion = parser.xpath("//div//div//h5[1]//strong/text()")
            descripcion = parser.xpath("//div//div//p[1]/text()")
            for destinos in nombre:
                lugar.append(destinos)
            
            for ubicacion_ in ubicacion:
                destino.append(ubicacion_)
            
            for descripcion_ in descripcion:
                detalles.append(descripcion_)
except:
    print(j)
```

```python
df = pd.DataFrame()
df["Nombre"] = pd.Series(lugar)
df["Destino"] = pd.Series(destino)
df["Descripcion"] = pd.Series(detalles)
```

```python
df.drop_duplicates().reset_index(drop = True)
```
